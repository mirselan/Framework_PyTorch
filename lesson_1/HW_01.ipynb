```python
import torch
import numpy as np
```

```python
from numpy.core.fromnumeric import size
# Создайте случайный FloatTensor размера 3x4x5
a = np.random.randint(0, 5, size=(3, 4, 5))
b = torch.FloatTensor(a)
print(a)
print(f'Размер: {a.shape}')
print(f'Тип: {a.dtype }')
print(b)
print(f'Размер: {b.size()}')
print("Количество измерений: %s" % b.dim())
print("Тип: %s\n" % (b.type()))
```

```python
# Выведите его форму (shape)
b.size()
```

```python
# Приведите его к форме 6 X 10
c = b.view(6, 10)
print(c)
print(c.size())
```

```python
# Умножьте его на вектор [1, 4, 2, 2, 1] поэлементно
vect = torch.tensor([1, 4, 2, 2, 1])
res = torch.mul(b, vect)
print(f'First variant:\n {res}\n')
print(f'Second variant:\n {b * vect}')
```

```python
# Умножьте тензор матрично на себя, чтобы результат был размерности 6x6
c_ = c.T
res_2 = torch.mm(c, c_)
print(res_2)
print(res_2.size())
```

```python
# Посчитайте производную функции y = x**3 + z - 75t в точке (1, 0.5, 2)
x=torch.tensor(1.0,requires_grad=True)
z=torch.tensor(0.5,requires_grad=True)
t=torch.tensor(2.0,requires_grad=True)
y=x**3 + z - 75*t

y.backward()  

print(x.grad)
print(z.grad)
print(t.grad)
```

```python
import sympy as sym
# create a "symbols"
x_ , z_, t_ = sym.symbols('x z t')
f = x_**3 + z_ - 75*t_

#Differentiating partially 
derivative_f_x = f.diff(x_)
derivative_f_z = f.diff(z_)
derivative_f_t = f.diff(t_)
print(derivative_f_x)
print(derivative_f_z)
print(derivative_f_t)
```
